```python

df1 = pd.pivot_table(d, index='REPORT_TYPE', values='ELEVATION', aggfunc='sum').reset_index()

# Printing the pivot table
df1
```

```python
report_type_counts = df1['REPORT_TYPE'].value_counts()

# Created the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=report_type_counts.index, y=report_type_counts.values)


plt.xlabel('Report Type')
plt.ylabel('Count')
plt.title('Count of Report Types')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
# Created a pivot table
df3 = pd.pivot_table(d, index='SOURCE', values='ELEVATION', aggfunc='sum').reset_index()

# Displayed the pivot table
print(df3)
```

```python
ax = sns.barplot(x='SOURCE', y='ELEVATION', data=d, estimator=sum)  # Using sum to aggregate ELEVATION values by SOURCE

# Setting the figure size
sns.set(rc={'figure.figsize':(12, 8.27)})

# Setting labels and title
ax.set(xlabel='Source', ylabel='Elevation', title='Source vs Elevation')

# Show the plot
plt.show()
```

```python
df5 = pd.pivot_table(d, index='CALL_SIGN', values='TMP', aggfunc='count').reset_index()

# Displaying the pivot table
print(df5)
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(df5['CALL_SIGN'], df5['TMP'], marker='o')  # Created the line plot with markers at each point
plt.title('Count of Temperature Recordings per Call Sign') 
plt.xlabel('Call Sign')  # X-axis label
plt.ylabel('Count of Temperature Recordings')  # Y-axis label
plt.grid(True)  # Added gridlines for better readability
plt.show()  # Display the plot
```
